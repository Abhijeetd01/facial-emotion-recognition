## Result analysys

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import brewer2mpl
from keras.backend import clear_session

from keras.callbacks import LambdaCallback, EarlyStopping,ModelCheckpoint
from keras.preprocessing.image import ImageDataGenerator

from models.final.model import myConvModel

from models.benchmark.model import BenchmarkModel

benchmark_model_weights="models/benchmark/my_model_weights.h5"

final_model_weights="models/final/my_model_weights.h5"

%matplotlib inline
```

## Load model

load both benchmark and final model and find the predictions

```python
X_test = np.load('data/X_test.npy')
y_test = np.load('data/y_test.npy')

print "X : ",X_test.shape
print "y : ",y_test.shape 

```

```python
benchmark=BenchmarkModel(benchmark_model_weights)
```

```python
y_pred_benchmark = benchmark.predict(X_test*(1.0/255.0))
y_pred_benchmark = np.argmax(y_pred_benchmark,1)
```

```python
clear_session() #clearing the session before loading the second model to avoid any conflict
mymodel=myConvModel(final_model_weights)
```

```python
y_pred = mymodel.predict(X_test*(1.0/255.0))
y_pred_onehot=y_pred
y_pred = np.argmax(y_pred,1)
```

## Accuracy

Accuracy is the most important metrics to compare our model to the benchmark 

Human Accuracy: 65.5%



```python
y_test=np.argmax(y_test,1)
benchmark_accuracy=1.0*sum(y_pred_benchmark==y_test)/len(y_test)
model_accuracy=1.0*sum(y_pred==y_test)/len(y_test)

print "Benchmark model accuracy:%.3f percent"%benchmark_accuracy
print "Final CNN model accuracy:%.3f percent"%model_accuracy
```

As we can see from the above result ,our model is providing much better prediction of emotion in comparison to our benchmark model. Also the accuracy score of our model is very much comparable to Human accuracy level

## Confusion matrix


```python
from sklearn.metrics import confusion_matrix
labels = ['angry', 'fear', 'happy', 'sad', 'surprise', 'neutral']


def plot_confusion_matrix(y_true, y_pred, cmap=plt.cm.Blues):
    cm = confusion_matrix(y_true, y_pred)
    i=0
    fig = plt.figure(figsize=(6,6))
    matplotlib.rcParams.update({'font.size': 16})
    ax  = fig.add_subplot(111)
    matrix = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    fig.colorbar(matrix) 
    for i in range(0,6):
        for j in range(0,6):  
            ax.text(j,i,cm[i,j],va='center', ha='center')
    # ax.set_title('Confusion Matrix')
    ticks = np.arange(len(labels))
    ax.set_xticks(ticks)
    ax.set_xticklabels(labels, rotation=45)
    ax.set_yticks(ticks)
    ax.set_yticklabels(labels)
    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

Benchmark model confusion matrix

```python
plot_confusion_matrix(y_test, y_pred_benchmark, cmap=plt.cm.YlGnBu)
```

Final model confusion matrix

```python
plot_confusion_matrix(y_test, y_pred, cmap=plt.cm.YlGnBu)
```

As we can see,for the benchmark model confusion matrix ,the model is predicting well for happy emotion ,but almost failing to predict all other emotions


But in our final model ,our model is pretty well predicting aalmost all labels correctly

## Failure case analysis


```python
set3 = brewer2mpl.get_map('Set3', 'qualitative', 6).mpl_colors
def plot_subjects(indexes, y_pred, y_true, title=False):
    fig = plt.figure(figsize=(12,12))
    emotion = {0:'Angry', 1:'Fear', 2:'Happy', 3:'Sad', 4:'Surprise', 5:'Neutral'}
    for i,indx in enumerate(indexes):
        input_img = X_test[indx].reshape((48,48))
        ax = fig.add_subplot(6,6,i+1)
        ax.imshow(input_img, cmap=matplotlib.cm.gray)
        plt.xticks(np.array([]))
        plt.yticks(np.array([]))
        if y_pred[indx] != y_true[indx]:
            plt.xlabel(emotion[y_true[indx]], color='#53b3cb',fontsize=12)
        else:
            plt.xlabel(emotion[y_true[indx]], fontsize=12)
        if title:
            plt.title(emotion[y_pred[indx]], color='blue')
        plt.tight_layout()
    plt.show()
    
def plot_probs(indexes, y_prob):
    fig = plt.figure(figsize=(12,12))
    for i,indx in enumerate(indexes):
        input_img = X_test[indx].reshape((48,48))
        ax = fig.add_subplot(6,6,i+1)
        ax.bar(np.arange(0,6), y_prob[indx], color=set3,alpha=0.5)
        ax.set_xticks(np.arange(0.5,6.5,1))
        labels = ['angry', 'fear', 'happy', 'sad', 'surprise','neutral']
        ax.set_xticklabels(labels, rotation=90, fontsize=10)
        ax.set_yticks(np.arange(0.0,1.1,0.5))
        plt.tight_layout()
    plt.show()
    
def plot_subjects_with_probs(indexes, y_prob):
    iter = len(indexes)/6
    for i in np.arange(0,iter):
        plot_subjects(indexes, y_pred, y_test, title=False)
        plot_probs(indexes, y_prob)
        
        
        
```

Showing some cases where the model failed to predict

```python
failed_cases = (y_test!=y_pred)
failed_indexes=[]
c=0
i=-1
while(c<18):
    i+=1
    if(failed_cases[i]):
        c+=1
        failed_indexes.append(i)
    
plot_subjects_with_probs(failed_indexes[0:6],y_pred_onehot)
plot_subjects_with_probs(failed_indexes[6:12],y_pred_onehot)
plot_subjects_with_probs(failed_indexes[12:18],y_pred_onehot)

```

As we can observe in most of the case even if the prediction failed for these cases,but the correct prediction matches mostly the second best prediction
